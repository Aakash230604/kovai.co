```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
def forecast_service(service_name, data, days=7):
    service_data = df[service_name]
    model = ARIMA(service_data, order=(1, 1, 1))
    model_fit = model.fit()
    forecast = model_fit.forecast(steps=days)
    return forecast
services=['Local Route','Light Rail','Peak Service','Rapid Route','School','Other']
```

```python
forecasts = {}
for service in services:
    forecasts[service] = forecast_service(service, df)
```

```python
for service, forecast in forecasts.items():
    print(f"Forecast for {service} for the next 7 days:")
    print(forecast)
```

```python
plt.figure(figsize=(12, 8))
for service, forecast in forecasts.items():
    plt.plot(forecast, label=service)
plt.legend()
plt.show()
```
